<h2>EDA: Multidimensional Data</h2>

<p>Not all Earth Systems Science data are stored in tables, though. In fact, a large percentage of the data used in the Earth Systems Sciences is array-based. While <code>pandas</code> is a great package for examining tabular data, it is insufficient for multidimensional or array-based data.</p>

<p>To examine these types of data we use a different package, <code>xarray</code>, with the abbreviated name <code>xr</code>.</p>

<p>The data being used this time are <i>remote</i>, meaning they are stored on a server that is external to the machine we're running this notebook. In this case, they are stored on a THREDDS Data Server, and we use the <code>siphon</code> Python package to bring it into our Python project.</p>



```python
# import required packages 
import xarray as xr
from siphon.catalog import TDSCatalog

# point to the catalog of interest on the THREDDS Data Server
cat = TDSCatalog('https://thredds.ucar.edu/thredds/catalog/grib/NCEP/GFS/Global_onedeg_ana/catalog.xml')
```

Then we use the <code>remote_access()</code> function from the <code>siphon</code> package to read the file, and load it into our Python project as an xarray <a href="https://docs.xarray.dev/en/stable/generated/xarray.Dataset.html" target="blank">DataSet</a>.

```python
# read the data from the remote TDSCatalog
ds = cat.datasets['Full Collection Dataset'].remote_access(use_xarray=True)

# preview the xarray dataSet
ds
```

Next in the EDA process, we will examine a single variable. In this case, temperature on isobaric surfaces. 

```python
# Isolate one data variable into a DataArray
# Note the syntax dataSet['variable']
# Use the tab key to help with variable names
temperature = ds['Temperature_isobaric']
temperature
```

<p>There are considerably more individual data points in this DataArray than there are in our previous time series data, so running calculations and simple statistics on all these data points will take too much time and hardware resources (causing out-of-memory errors). Let's continue our EDA by selecting just one time to examine, reducing the dimensionality from 4 (x, y, z, and t) to just 3 (x, y, and z). </p>

<p>To select a single time, we use the <code>.sel()</code> method in the <code>xarray</code> package, giving it a single time as input.</p>

<p>Consider this: <i>where can you find a valid time for this Data Array?</i></p>

```python
# Note the syntax DataArray.sel(dimension='value')
temperature_selection = temperature.sel(time='2025-06-03T06')
temperature_selection
```

```python
# examine the range of all temperature values at this time
temperature_selection.plot()
```

```python
# find nulls in temperature
temperature_selection.isnull().sum()
```

<p>Multiple selections can be made at once by adding additional constraints to the <code>.sel()</code> method. Take for example the relative humidity at 1000 hPa (100000 Pa) for a single time step.</p>

<h2>Poster-Ready Plotting</h2>

<p>We're skipping a few steps in the scientific process, but after performing an exploratory data analysis, running calculations and analyses, you're likely to need to visually present your data on a plot.</p>

<p>Data come in all shapes and sizes, and so do the resulting visualizations. In our short time together, we'll explore a few different types of plots that can be created in the scientific Python ecosystem. However, very importantly, do not expend too much of your energy trying to commit these lines of code to memory. An important skill to learn when plotting figures in Python is <b><i>copy+paste</i></b>!</p>

<div class="alert alert-info">
    <b>Tip:</b>
    <p>There are plenty of <a href="https://matplotlib.org/cheatsheets/" target="blank">cheat sheets</a> out there to help on your copy+paste adventures.</p>

<p>One of the most commonly used Python plotting packages is <a href="https://matplotlib.org/" target="blank">matplotlib</a>. Let's begin building our foundational Python plotting knowledge by recognizing a few important components of a matplotlib figure.</p> 

<p>All elements of a plot are contained within a figure object, which may contain one or many axes objects (or none at all, but that doesn't make a very useful figure!). Axes display data.</p>

<p>When we created our earlier plots, such as <code>df[col].plot()</code> and <code>rh_selection.plot()</code>, the figure and axes objects were <i>implicitly</i> created for us behind the scenes. This is very handy for EDA plots, but we need more control for poster-ready plots.</p>

<p>Figures and axes have a number of properties associated with them that can be modified. For poster-ready plots, we <i>explicitly</i> create these objects so we can manipulate their default values. To create these objects, we need to import matplotlib's pyplot interface, <code>matplotlib.pyplot</code>. You will commonly see this imported with the "plt" abbreviation as a shortcut.</p>

<p>In the next section we'll briefly touch on a variety of different plot types available in <code>matplotlib</code>. </p>

<h3>Plot Potpourri (Plotpourri?)</h3>

```python
# import matplotlib's plotting interface
import matplotlib.pyplot as plt

# import numpy to create simple, demonstration data
import numpy as np

# create demo data to showcase the different kinds of plots we can make
x = np.linspace(0.5, 3.5, 100)
y1 = 3+np.cos(x)
y2 = 1+np.cos(1+x/0.75)/2
```

<h4>Line plot</h4>

```python
# create figure and axes
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1, 1, 1)  # you can leave this blank for single subplots

# plot data on axes, "ax"
ax.plot(x, y1, color='blue', linewidth=2.0, linestyle='dashed', label='y2')
ax.plot(x, y2, 'm+', label='y2')

# set extent of axes
ax.set_xlim(0, 3)
ax.set_ylim(0.5, 3.5)

# set labels
ax.set_ylabel('my y-axis')
ax.set_xlabel('my x-axis')

# set title
ax.set_title('Hey, this is my plot!')

# add legend
ax.legend()
```

<h4>Bar Chart</h4>

```python
# create demo data
regions = ['Pacific Northwest', 'Southeast', 'New England']
counts = [12, 5, 45]
colors = ['red', 'brown', 'orange']
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.bar(regions, counts, color=colors, alpha=0.8)  # can also just choose one color for them all!

ax.set_title('Days With Air Quality Warning Days in 2023')
```

<h4>Scatter plot</h4>

```python
# create demo data
y3 = np.random.uniform(1, 3, len(x))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

s = ax.scatter(x, y3, c=y1*100, cmap='viridis')

cb = fig.colorbar(s, label='Temperature [K]', extend='both')

ax.set_yticks(np.arange(1, 3.5, 0.5))
ax.set_xticks(range(0, 5))

ax.grid()

ax.set_xlabel('x-coordinates [m]')
ax.set_ylabel('y-coordinates [yds]')

ax.set_title('Scattered points, with more data!')

# save the figure to a png to use in your poster!
plt.savefig('scatterplot.png')
```

<h2>Geographic Plots</h2>

<p>Weather and climate occur on a globe, a 3D surface. Atmospheric and climate scientists monitor weather and climate primarily through digital screens, 2D surfaces. Hopefully, none of that is news to you! But what you may have not considered are the inherent distortions that our meteorological data have when we display it on 2D surfaces. Have you heard of the <a href="https://www.youtube.com/watch?v=LN_tCKdl9cE" target="blank">orange peel problem</a>? These distortions led to the need for <i>coordinate reference systems</i> for geographic data. </p>

<p>A coordinate reference system refers to a generic system for referencing geographic data onto a surface. We can have projected coordinate reference systems as well as geographic coordinate reference systems, but this lesson will primarily refer to projected coordinate reference systems, or "projections". Cartographers over the years have created several projections or methods to stretch and distort 3D surfaces onto a 2D surface. Projections can preserve attributes like distance, direction, shape, or area, but never all four.</p>

<p>If we want to display geographically realistic depictions of our meteorological data, we need to plot our data not on a Cartesian grid (such as those offered by matplotlib functions), but instead on axes that take a projection into account. This is where <code>Cartopy</code> comes in. </p>

<p><a href="https://scitools.org.uk/cartopy/docs/latest/" target="blank">Cartopy</a> builds on matplotlib plotting functions to offer projected axes to plot geographic data on. Cartopy offers a variety of <a href="https://scitools.org.uk/cartopy/docs/latest/reference/crs.html#list-of-projections" target="blank">projections</a> as well as geographic reference features that make adding context to maps simple.</p>

<p>Creating a figure with geographic axes is very similar to creating any other plot with Matplotlib. First import Cartopy's projection engine <code>ccrs</code> and matplotlib using the standard conventions.</p>

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

<p>Then create a figure and axes, but include the <code>projection</code> keyword in the <code>add_subplot()</code> method. Here's an example of how to create axes with the Lambert Conformal projection.</p>

```python
fig = plt.figure()
ax = fig.add_subplot(projection=ccrs.LambertConformal())

# add Cartopy's built-in coastlines for reference
ax.coastlines()
```

<p>It's important to know which reference system your data are recorded in so that they can be matched to the correct location on the geographic plot. MetPy provides a function for easily getting that information, provided that the data uses following <a href="https://cfconventions.org/" target="blank">CF conventions</a>, called <code>parse_cf()</code>. Let's try this on the NCEP Analysis output from earlier.</p>

```python
import metpy

# execute parse_cf function
ds = ds.metpy.parse_cf()

# display the new coordinate
ds.metpy_crs
```

<p>The coordinate reference system in this case uses latitude and longitude. While this is technically not a "projected" coordinate reference system, we can simulate this coordinate reference system as a projection using the Plate Carrée projection. The <a href="https://pro.arcgis.com/en/pro-app/latest/help/mapping/properties/plate-carree.htm" target="blank">Plate Carrée projection</a> is commonly used in global model output, as it displays data in equally spaced latitudes and longitudes and preserves direction.</p>

<p>Now that we have this information, we can use this to plot any 2D variables from the NCEP analysis DataSet. We will need the following:</p>

<ol>
    <li>2D data array</li>
    <li>the x coordinates corresponding to each cell in the 2D array</li>
    <li>the y coordinates corresponding to each cell in the 2D array</li>
    <li>the coordinate reference system of the data</li>
    <li>the desired coordinate reference system for the resulting plot</li>
</ol>

<p>Since we used the <code>parse_cf()</code> method earlier, we can use MetPy shortcuts to find items 1-3. We just found item 4 above, and item 5 can be anything we choose!</p>

```python
# 1. 2D data array
data = ds['Relative_humidity_isobaric'].sel(time='2025-06-03T06', isobaric=100000)

# 2 and 3. x and y coordinates
lon = data.metpy.x
lat = data.metpy.y

# 4. crs of the data
data_crs = ccrs.PlateCarree()

# 5. crs of the plot
plot_crs = ccrs.LambertConformal()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection=plot_crs)

# display the data using filled contours, contourf()
cntrf = ax.contourf(lon, lat, data, transform=data_crs, cmap='coolwarm')
cb = fig.colorbar(cntrf, label='Relative Humidity (%)', extend='both')

# set the axes extent in the format [western border, eastern border, southern border, northern border]
ax.set_extent([-121, -68, 10, 70], crs=ccrs.PlateCarree())

# add Cartopy's built-in coastlines for reference
ax.coastlines()
```
